```python
from sklearn.preprocessing import StandardScaler
```

Insights :

-  To normalize data we use standard scaler.

```python
scaler=StandardScaler()
```

```python
scaler.fit(df.drop(['type','animal name'],axis=1))
```

```python
scaled_features=scaler.transform(df.drop(['type','animal name'],axis=1))
df_head=pd.DataFrame(scaled_features,columns=df.columns[1:-1])
```

```python
df.columns[1:-1]
```

Insights : 

-  While doing satndard scaler in above syntax in scaler.transform we drop two column which is Type,animal name.

-  Storing that transformation in dataframe (df_head) excluding colums 'type' and 'animal name', which is first and last column in dataframe.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_head,df['type'],test_size=0.3,random_state=40)
```

Insight :

-  With the help of train_test split we divided our dataframe into two parts (train and test)
-  we divided our data into train and test data into 70-30 size means our train data size is 70% of the whole data and test data size is 30% of the whole data
-  we divided our data in two data set because we perform all EDA, visualization, and taking some insights from train data, and all insights which is conclued from that data will apply on test data to predict and to check how good our model is work

KNN Classifier

KNN - 

-  It classifies the data points on how its neighbor is classfied.

-  classifies based on the similarity measure of the earlier stored data points.


```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(x_train,y_train)
```

Insight :

-  .fit is use to fit data in model 
-  Here we are provide training data to model for fiting purpose

```python
pred = KNN.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

Insights :

-  Here confusion matrix contains the number of correct and incorrect predictions of the classification model

-  Classification report is a performance evaluation metric. it is use to show precision, recall and F1 score.

```python
print('Classification Report',classification_report(y_test,pred))
```

```python
print('Confusion Matrix',confusion_matrix(y_test,pred))
```

Insights : 

-  Here we are using Elbow method to choose the best valur for K

```python
error_rate=[]
for i in range(1,40):
    KNN=KNeighborsClassifier(n_neighbors=i)
    KNN.fit(x_train,y_train)
    pred_i=KNN.predict(x_test)
    error_rate.append(np.mean(pred_i!=y_test))
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
plt.title=('Error_rate vs K value')
plt.xlabel('K')
plt.ylabel('Error rate')
```

Insights :

-  From the above diagram we can say that k=1 have low error rate.

-  Hence we will perform the test with k=1

```python
KNN = KNeighborsClassifier(n_neighbors=1)
```

```python
KNN.fit(x_train,y_train)
```

```python
pred = KNN.predict(x_test)
```

```python
print('Classification Report',classification_report(y_test,pred))
```

Insights : 

-  We can clearly see that after giving k=1 i.e. proper value of k our classification report is improved.

```python
print('Confusion Matrix'),confusion_matrix(y_test,pred)
```

Insights :

-  We can see that at k=1 precision has increases. Hence selecting k = 1 is justified.
